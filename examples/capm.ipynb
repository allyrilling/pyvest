```python
from pyvest import FamaFrenchDataReader, CAPM
```

# Fama-French Data Reader

```python
# Define dates
start_date = "1964-01-31"
end_date = "2023-08-31"

# Define the table to extract
beta_data_set = 'Portfolios_Formed_on_BETA'
me_beta_data_set = "25_Portfolios_ME_BETA_5x5"
```

```python
data_reader = FamaFrenchDataReader()
```

## Read factors

```python
factors = data_reader.read_factors(start_date, end_date)
factors
```

## Get list of available datasets

```python
data_reader.get_available_datasets()
```

## Read portfolio returns

```python
beta_portfolios = data_reader.read_returns(beta_data_set, start_date, end_date)
beta_portfolios
```

```python
me_beta_portfolios = data_reader.read_returns(me_beta_data_set, start_date, end_date)
me_beta_portfolios
```

# Create CAPM model

```python
beta_capm = CAPM(factors, beta_portfolios, name="BETA")
```

```python
me_beta_capm = CAPM(factors, me_beta_portfolios, name="ME-BETA")
```

```python
beta_capm.name
```

Note that the **plot()** method calculates automatically the regressions. In other words, you never really have to call **calculate_regressions()** explicitly.
